In this notebook I estimate the correlation of the RBM response scores to the clonal fold change in in vitro T-cell stimulation experiments from Balachandran et al. 2017. The related figures are 3 and S3, S6.

```python
rootf = FOLDER ## place here the folder where you save all the folders of the git ##

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = ['Garuda']
mpl.rcParams['font.serif'] = ['Garuda-Oblique']

curr_float = np.float32
curr_int = np.int16

def convert_number(seqs): # convert to numbers already aligned seqs
    aa = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V',  'W', 'Y','-']
    aadict = {aa[k]: k for k in range(len(aa))}
    msa_num = np.array(list(map(lambda x: [aadict[y] for y in x], seqs[0:])), dtype=curr_int, order="c")
    
    return msa_num

def convert_letter(seqs_n): # convert to numbers already aligned seqs
    aa = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V',  'W', 'Y','-']
    aadictinv = {k: aa[k] for k in range(len(aa))} 
    seqs=[]
    if type(seqs_n[0]) == curr_int:
        seqs.append(''.join([aadictinv[e] for e in seqs_n]))
    else:
        for t in range(len(seqs_n)):
            seqs.append(''.join([aadictinv[e] for e in seqs_n[t]]))
    return seqs

def uniqueIndexes(l):
    seen = set()
    res = []
    for i, n in enumerate(l):
        if n not in seen:
            res.append(i)
            seen.add(n)
    return res

#%matplotlib inline
import sys, os, pickle
sys.path.append('/home/barbara/.local/lib/python3.6/site-packages/')
sys.path.append(rootf + '/PGM/source/')
sys.path.append(rootf + '/PGM/utilities/')
sys.path.append(rootf + '/Align_utils/')
from common_imports import set_num_threads
set_num_threads(1) # Set the number of cores. Must be executed before importing numpy&numba.
import rbm,utilities
import Proteins_utils, RBM_utils, utilities, sequence_logo, plots_utils

def countX(lst, x): 
    return lst.count(x) 

listpep1 = ['WT', 'NA', 'MUC16ID', 'MUC16WT', 'CR'];
listpep2 = ['WT', 'NA', 'MUC16ID', 'MUC16WT', 'CR'];
listpep3 = ['NA', 'CR'];
listpep4 = ['WT', 'NA', 'CR'];
listpep5 = ['NA', 'CR'];
listpep6 = ['NA1', 'NA2', 'CR1', 'CR2'];
listpep7 = ['NA','CR']; 
listpep=[listpep1,listpep2,listpep3,listpep4,listpep5,listpep6,listpep7];
```

Here I evaluate the likelihood for the CDR3 sequences in the validation set for each sample under 
the RBM model and its Left-Right version  

```python
path_in = rootf + '/Data/Count-weighted'

amino_acids = 'ACDEFGHIKLMNPQRSTVWY'
aadict = {amino_acids[k]: k for k in range(len(amino_acids))} 

max_L = 19
max_depth = max_L

features = []
for aa in amino_acids:
    features += [['a' + aa + str(L)] for L in range(max_depth)]
    features += [['a' + aa + str(L)] for L in range(-max_depth, 0)] # features for encoding

features_r = []
for aa in amino_acids:
    features_r += [['a\t' + aa + '\t' + str(L)] for L in range(max_depth)]
    features_r += [['a\t' + aa + '\t' + str(L)] for L in range(-max_depth, 0)]  # features for writing

feature_dict = {tuple(f): i for i, f in enumerate(features)}
feature_dict_inv = {i:f for i, f in enumerate(features)}

n_v = 19 # Number of visible units; = # sites in alignment.
n_h = 25 # Number of hidden units.
hu = n_h

l12 = 0.1
TS = 80

listcq=[]
listcp=[]
listcq_lr=[]
listcp_lr=[]
frac_all=[]
tot_all=[]
listcq_pgen=[]
listcq_ppost=[]

for pp in range(1,8):
    for r in range(len(listpep[pp-1])):
        pep=listpep[pp-1][r]
        ptn=pp
        
        summary = pd.read_csv(rootf + '/Data/summary_pt'+str(ptn)+'_' + pep + '.tsv', sep='\t', low_memory=False)
        seqs_gapless0 = list(summary['CDR3'])
        subset = uniqueIndexes(seqs_gapless0)
        summary1=summary.loc[subset,:]
        seqs=list(summary1['CDR3_gap'].values)
        seqs_gapless=list(summary1['CDR3'].values)

        tt='day0'
        path_od0 = rootf + '/Models/CDR3out_pt' + str(ptn) + '_' + str(tt) 
        filename_day0 = '/CDR3fasta_pt' + str(ptn) + '_' + str(tt) +'.fasta'
        all_data_day0 = Proteins_utils.load_FASTA(path_in + filename_day0,with_labels=False,remove_insertions=False, drop_duplicates=False)
        Bd0 = all_data_day0.shape[0] # here I set the weight of each seqs to 1
        
        tt='3w'
        path_o3w = rootf + '/Models/CDR3out_pt' + str(ptn) + '_' + pep + '_' + str(tt) 
        filename_3w = '/CDR3fasta_pt' + str(ptn) + '_' + pep +'_' + str(tt) +'.fasta'
        all_data_3w = Proteins_utils.load_FASTA(path_in + filename_3w,with_labels=False,remove_insertions=False, drop_duplicates=False)
        B3w = all_data_3w.shape[0] # here I set the weight of each seqs to 1
          
        data_val_3w_n = all_data_3w[int(B3w*TS/100):]
        data_val_day0_n = all_data_day0[int(Bd0*TS/100):]
        data_val_3w = convert_letter(data_val_3w_n)
        data_val_day0 = convert_letter(data_val_day0_n)

        mult3w=[]
        multd0=[]
        seqs_val=[]
        for ss in seqs:
            c3=countX(data_val_3w, ss)
            c0=countX(data_val_day0, ss)
            if c0 or c3:
                mult3w.append(c3)
                multd0.append(c0)
                seqs_val.append(ss)
                
        mult3w=np.array(mult3w)
        multd0=np.array(multd0)
        seqs_val_n = convert_number(seqs_val)
        
        namem_d0 = path_od0 + '/model_' + str(hu) + '_' + str(l12) + '.data'
        namem_3w = path_o3w + '/model_' + str(hu) + '_' + str(l12) + '.data'
        RBM_d0 = RBM_utils.loadRBM(namem_d0)
        RBM_3w = RBM_utils.loadRBM(namem_3w)
        llRBM = RBM_3w.likelihood(seqs_val_n)
        llRBM0 = RBM_d0.likelihood(seqs_val_n)
        
        RBM_pgen = RBM_utils.loadRBM(rootf + '/Models/CDR3out_Pgen/model_' + str(hu) + '_' + str(l12) + '.data')
        RBM_post = RBM_utils.loadRBM(rootf + '/Models/CDR3out_Ppost/model_' + str(hu) + '_' + str(l12) + '.data')
        llRBMgen = RBM_pgen.likelihood(seqs_val_n)
        llRBMpost = RBM_post.likelihood(seqs_val_n)

        mp1 = np.log10((mult3w+1/2)/(multd0+1/2))
        mp = np.log10((mult3w+1/2))
        listc_p=[]
        listc_q=[]
        listc_pgen=[]
        listc_ppost=[]
        frac=[]
        tot=[]
        for th in list(np.arange(-1,2.2,0.1)):
            listc_p.append(np.corrcoef(llRBM[mp>th], mp[mp>th])[0,1])
            listc_q.append(np.corrcoef((llRBM-llRBM0)[mp>th], mp1[mp>th])[0,1])
            listc_pgen.append(np.corrcoef((llRBM-llRBMgen)[mp>th], mp[mp>th])[0,1])
            listc_ppost.append(np.corrcoef((llRBM-llRBMpost)[mp>th], mp[mp>th])[0,1])
            
            frac.append(len(mp[mp>th])/len(mp))
            tot.append(len(mp[mp>th]))
        frac_all.append(frac)
        tot_all.append(tot)
        
        listcq.append(listc_q) # List of correlation between response score and fold change for RBM
        listcq_pgen.append(listc_pgen) # List of correlation between response score and fold change for RBM - response score calculated using Pgen as background
        listcq_ppost.append(listc_ppost) # List of correlation between response score and fold change for RBM - response score calculated using Ppost as background
        listcp.append(listc_p) # List of correlation between log-likelihood and abundance for RBM
        
        # From here the Left-Right version #
        data_seqs_val_3w = []
        for t in range(len(seqs_val)):
            ij = seqs.index(seqs_val[t])
            seq = seqs_gapless[ij]
            seq_feature_lsts = [['a' + aa + str(i)] for i, aa in enumerate(seq)]
            seq_feature_lsts += [['a' + aa + str(-1-i)] for i, aa in enumerate(seq[::-1])]
            seq_features = list(set([feature_dict[tuple(f)] for f in seq_feature_lsts if tuple(f) in feature_dict]))
            length_input=len(features)
            data=np.array(seq_features)
            data_enc = np.zeros(length_input, dtype=np.int8)
            for i in range(len(data)): data_enc[data[i]] = 1
            data_seqs_val_3w.append(list(data_enc))  

        seqs_val_n = np.array(data_seqs_val_3w).astype(np.int16)
        
        namem_d0 = path_od0 + '/model_' + str(hu) + '_' + str(l12) + '_LR.data'
        namem_3w = path_o3w + '/model_' + str(hu) + '_' + str(l12) + '_LR.data'
        RBM_d0 = RBM_utils.loadRBM(namem_d0)
        RBM_3w = RBM_utils.loadRBM(namem_3w)
        llRBM = RBM_3w.likelihood(seqs_val_n)
        llRBM0 = RBM_d0.likelihood(seqs_val_n)
        
        listc_p=[]
        listc_q=[]
        frac=[]
        tot=[]
        for th in list(np.arange(-1,2.2,0.1)):
            listc_p.append(np.corrcoef(llRBM[mp>th], mp[mp>th])[0,1])
            listc_q.append(np.corrcoef((llRBM-llRBM0)[mp>th], mp1[mp>th])[0,1])
        
        listcq_lr.append(listc_q) # List of corrlation between response score and fold change for RBM (Left+Right format)
        listcp_lr.append(listc_p) # List of corrlation between log-likelihood and abundance for RBM (Left+Right format)
        
# Example for writing #
writer = pd.ExcelWriter(rootf + '/Data/summary_response_score_rbm.xlsx', engine = 'xlsxwriter')
ss=0
model = 'RBM'
for pp in range(1,8):
    for r in range(len(listpep[pp-1])):
        pep=listpep[pp-1][r]
        ptn=pp
        datao = {'Dataset size (filtered by counts)': tot_all[ss],
                'Correlation response score - fold change': listcq[ss],
                'Correlation to abundance (day 21)': listcp[ss],
                'Correlation response score (Pgen) - fold change': listcq_pgen[ss],
                'Correlation response score (Ppost) - fold change': listcq_ppost[ss]}

        df = pd.DataFrame(datao, columns = ['Dataset size (filtered by counts)','Correlation response score - fold change','Correlation to abundance (day 21)','Correlation response score (Pgen) - fold change','Correlation response score (Pppost) - fold change'])
        df.to_excel(writer, sheet_name = model + '_pat_' + str(pp) + '_' + pep, index=0)
        ss+=1
    
ss = 0
model = 'RBM_LR'
for pp in range(1,8):
    for r in range(len(listpep[pp-1])):
        pep=listpep[pp-1][r]
        ptn=pp
        datao = {'Dataset size (filtered by counts)': tot_all[ss],
                'Correlation response score - fold change': listcq_lr[ss],
                'Correlation to abundance (day 21)': listcp_lr[ss]}

        df = pd.DataFrame(datao, columns = ['Dataset size (filtered by counts)','Correlation response score - fold change','Correlation to abundance (day 21)'])
        df.to_excel(writer, sheet_name = model + '_pat_' + str(pp) + '_' + pep, index=0)
        ss+=1
        
writer.save()
```
