```python
#rootf = FOLDER ## place here the folder where you save all the folders of the git ##
rootf = '/home/barbara/Barbara_Bravi/rbm'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = ['Garuda']
mpl.rcParams['font.serif'] = ['Garuda-Oblique']

curr_float = np.float32
curr_int = np.int16

def convert_number(seqs): # convert to numbers already aligned seqs
    aa = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V',  'W', 'Y','-']
    aadict = {aa[k]: k for k in range(len(aa))}
    msa_num = np.array(list(map(lambda x: [aadict[y] for y in x], seqs[0:])), dtype=curr_int, order="c")
    
    return msa_num

def convert_letter(seqs_n): # convert to numbers already aligned seqs
    aa = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V',  'W', 'Y','-']
    aadictinv = {k: aa[k] for k in range(len(aa))} 
    seqs=[]
    if type(seqs_n[0]) == curr_int:
        seqs.append(''.join([aadictinv[e] for e in seqs_n]))
    else:
        for t in range(len(seqs_n)):
            seqs.append(''.join([aadictinv[e] for e in seqs_n[t]]))
    return seqs

def uniqueIndexes(l):
    seen = set()
    res = []
    for i, n in enumerate(l):
        if n not in seen:
            res.append(i)
            seen.add(n)
    return res

#%matplotlib inline
import sys, os, pickle
sys.path.append('/home/barbara/.local/lib/python3.6/site-packages/')
sys.path.append(rootf + '/PGM/source/')
sys.path.append(rootf + '/PGM/utilities/')
sys.path.append(rootf + '/Align_utils/')
from common_imports import set_num_threads
set_num_threads(1) # Set the number of cores. Must be executed before importing numpy&numba.
import rbm,utilities
import Proteins_utils, RBM_utils, utilities, sequence_logo, plots_utils
```

```python
# RBM recovers sequence-counts correlation: 
# we evaluate the response score on the validation set and calculate the correlation to the fold change filtering 
# progressively out the clones at high abundance

path_in = rootf + '/Data/CDR3'

n_v = 19 # Number of visible units; = # sites in alignment.
n_h = 25 # Number of hidden units.
hu = n_h

l12 = 0.1
TS = 80
import pandas as pd

def countX(lst, x): 
    return lst.count(x) 

listpep1 = ['WT', 'NA', 'MUC16ID', 'MUC16WT', 'CR'];
listpep2 = ['WT', 'NA', 'MUC16ID', 'MUC16WT', 'CR'];
listpep3 = ['NA', 'CR'];
listpep4 = ['WT', 'NA', 'CR'];
listpep5 = ['NA', 'CR'];
listpep6 = ['NA1', 'NA2', 'CR1', 'CR2'];
listpep7 = ['NA','CR']; 
listpep=[listpep1,listpep2,listpep3,listpep4,listpep5,listpep6,listpep7];

listcq=[]
listcp=[]
frac_all=[]
tot_all=[]
for pp in range(1,8):
    for r in range(len(listpep[pp-1])):
        pep=listpep[pp-1][r]
        ptn=pp
        
        summary = pd.read_csv(rootf + '/Data/summary_pt'+str(ptn)+'_' + pep + '.tsv', sep='\t', low_memory=False)
        seqs = list(summary['CDR3'])
        subset = uniqueIndexes(seqs)
        summary1=summary.loc[subset,:]
        seqs=list(summary1['CDR3_gap'].values)

        tt='day0'
        path_od0 = rootf + '/Models/CDR3out_pt' + str(ptn) + '_' + str(tt) 
        filename_day0 = '/CDR3fasta_pt' + str(ptn) + '_' + str(tt) +'.fasta'
        all_data_day0 = Proteins_utils.load_FASTA(path_in + filename_day0,with_labels=False,remove_insertions=False, drop_duplicates=False)
        Bd0 = all_data_day0.shape[0] # here I set the weight of each seqs to 1
        
        tt='3w'
        path_o3w = rootf + '/Models/CDR3out_pt' + str(ptn) + '_' + pep + '_' + str(tt) 
        filename_3w = '/CDR3fasta_pt' + str(ptn) + '_' + pep +'_' + str(tt) +'.fasta'
        all_data_3w = Proteins_utils.load_FASTA(path_in + filename_3w,with_labels=False,remove_insertions=False, drop_duplicates=False)
        B3w = all_data_3w.shape[0] # here I set the weight of each seqs to 1
          
        data_val_3w_n = all_data_3w[int(B3w*TS/100):]
        data_val_day0_n = all_data_day0[int(Bd0*TS/100):]
        data_val_3w = convert_letter(data_val_3w_n)
        data_val_day0 = convert_letter(data_val_day0_n)

        mult3w=[]
        multd0=[]
        seqs_val=[]
        for ss in seqs:
            c3=countX(data_val_3w, ss)
            c0=countX(data_val_day0, ss)
            if c0 or c3:
                mult3w.append(c3)
                multd0.append(c0)
                seqs_val.append(ss)
                
        mult3w=np.array(mult3w)
        multd0=np.array(multd0)
        seqs_val_n = convert_number(seqs_val)
        
        namem_d0 = path_od0 + '/model_' + str(hu) + '_' + str(l12) + '.data'
        namem_3w = path_o3w + '/model_' + str(hu) + '_' + str(l12) + '.data'

        RBM_d0 = RBM_utils.loadRBM(namem_d0)
        RBM_3w = RBM_utils.loadRBM(namem_3w)
        we_d0 = RBM_d0.weights
        we_3w = RBM_3w.weights

        llRBM = RBM_3w.likelihood(seqs_val_n)
        llRBM0 = RBM_d0.likelihood(seqs_val_n)
        
        mp1 = np.log10((mult3w+1/2)/(multd0+1/2))
        mp = np.log10((mult3w+1/2))
        listc_p=[]
        listc_q=[]
        frac=[]
        tot=[]
        for th in list(np.arange(-1,2.2,0.1)):
            listc_p.append(np.corrcoef(llRBM[mp>th], mp[mp>th])[0,1])
            listc_q.append(np.corrcoef((llRBM-llRBM0)[mp>th], mp1[mp>th])[0,1])
            
            frac.append(len(mp[mp>th])/len(mp))
            tot.append(len(mp[mp>th]))
        frac_all.append(frac)
        tot_all.append(tot)
        
        listcq.append(listc_q)
        listcp.append(listc_p)
```

```python
## visualize the correlations for a sample (Pt3 NA as in figures ...)
ss = 10 ## index for pt3 NA ##
point = np.argmin(abs(np.array(tot_all[ss]) - 10**(2.1)))
s2=16
fig, ax=plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(5)
fig.subplots_adjust(bottom = 0.4)
plt.plot(tot_all[ss],listcq[ss], linewidth = 2)
ax.set_ylabel(r'Correlation $\mathcal{S}_{\rm resp}^P$-fold change', fontsize=s2)
ax.set_xlabel('Dataset size (filtered by counts)', fontsize=s2)
plt.suptitle('Dataset: Pt3 NA', fontsize=s2-2)
sc=80
ax.scatter(tot_all[ss][point], listcq[ss][point], color = 'DarkGoldenRod', zorder=6, s=sc, edgecolors='k')
plt.xscale('log')
#xinf=0.006
xinf=60
ax.set_xlim([xinf,10000])
ax.set_ylim([0.,1.])
ax.tick_params(axis='both', which='major', labelsize = s2)
```

```python
# Summary plots over all the samples

list_pat=[]
list_pep=[]
list_corr001=[] 
ff=0
fracs=[]
for pp in range(1,8):
    for r in range(len(listpep[pp-1])):
        pep = listpep[pp-1][r]
        
        list_pat.append(pp)
        list_pep.append(pep)
        
        point = np.argmin(abs(np.array(tot_all[ff]) - 10**(2.1)))
        fracs.append(frac_all[ff][point])
        list_corr001.append(listcq[ff][point])
        
        ff+=1
        
## without background - almost unchanged ##
name_tech = u'Correlation RBM $\mathcal{S}_{resp}^P$-fold change'

import pandas as pd 

datao = {'Patient': list_pat,
        'Peptide': list_pep,
        name_tech: list_corr001,
        }

df = pd.DataFrame (datao, columns = ['Patient','Peptide', name_tech])

list_NA = ['NA', 'NA1', 'NA2']
list_MUC16 = ['MUC16ID']
list_XRM = ['CR', 'CR1', 'CR2']
list_WT = ['WT']
list_MUC16WT = ['MUC16WT']

auroc_xrm = list(df[df['Peptide'].isin(list_XRM)][name_tech].values)
auroc_wt_muc = list(df[df['Peptide'].isin(list_WT)][name_tech].values) + list(df[df['Peptide'].isin(list_MUC16WT)][name_tech].values)
auroc_na_muc = list(df[df['Peptide'].isin(list_NA)][name_tech].values) + list(df[df['Peptide'].isin(list_MUC16)][name_tech].values)

names_na_muc = ['Pt1 NA','Pt2 NA','Pt3 NA','Pt4 NA','Pt5 NA','Pt6 NA1', 'Pt6 NA2', 'Pt7 NA', 'Pt1 MUC16','Pt2 MUC16']
names_na = names_na_muc[:-2]
names_wt_muc=['Pt1 WT','Pt2 WT','Pt4 WT', 'Pt1 MUC16WT','Pt2 MUC16WT']
names_wt = names_wt_muc[:-2]
names_xrm=['Pt1 CR','Pt2 CR','Pt3 CR','Pt4 CR','Pt5 CR','Pt6 CR1', 'Pt6 CR2', 'Pt7 CR']

colors3 = list(np.repeat(['DarkGoldenRod'],len(names_na_muc))) + list(np.repeat(['Maroon'],len(names_wt_muc))) + list(np.repeat(['CadetBlue'],len(names_xrm)))
s2=16
sc=2
fig,ax=plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(12)
fig.subplots_adjust(bottom = 0.4)
meth = names_na_muc + names_wt_muc + names_xrm
meth_val = np.array(auroc_na_muc + auroc_wt_muc + auroc_xrm) 
x_pos = np.arange(len(meth_val))
x_pos = np.arange(0,2*len(meth_val),2)
ax.bar(x_pos, meth_val, alpha=1, edgecolor='k', align = 'center', capsize=sc, color = colors3)
ax.set_ylim(0.1, 1)
ax.set_xticks(x_pos)
ax.set_ylabel(name_tech, fontsize = 14)
ax.tick_params(axis='both', which='major', labelsize = 14)
ax.set_xticklabels(meth, rotation = 30, fontsize = 12, ha='right', rotation_mode = 'anchor');
```
